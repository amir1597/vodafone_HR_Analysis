```python
dfs = [af , central_usa ,cse ,lat_pa ,mea ,mer ,Subsaharan,F_east]
vd_all_loc = pd.concat(dfs, ignore_index=True)
```

```python
vd_all_loc1 = vd_all_loc.copy()
```

```python
vd_all_loc1 = vd_all_loc1.melt(id_vars=['DIM', 'Task', 'APPROVAL', 'DEP'], 
                               var_name='Date', 
                               value_name='Value')
vd_all_loc1
```

```python
transformed_data = []
employee = None
contracted_hours = None
allocated_hours = None
remaining_hours = None
current_date = None

for index, row in vd_all_loc1.iterrows():
    if isinstance(row['DIM'], str) and 'EMP_' in row['DIM']:  
        if employee is not None:
            transformed_data.append([employee, contracted_hours, allocated_hours, remaining_hours, current_date])
        
        employee = row['DIM']
        current_date = row['Date']  # Retain the date
        contracted_hours = None
        allocated_hours = None
        remaining_hours = None
    
    elif row['DIM'] == 'Contracted Hours':
        contracted_hours = row['Value']
    
    elif row['DIM'] == 'Allocated Hours':
        allocated_hours = row['Value']
    
    elif row['DIM'] == 'Remaining':
        remaining_hours = row['Value']

# Step 3: Append the last employee's data
transformed_data.append([employee, contracted_hours, allocated_hours, remaining_hours, current_date])

# Step 4: Create a new DataFrame from the transformed data, including the Date column
vd_all_loc1_transformed = pd.DataFrame(transformed_data, columns=['Employee Code', 'Contracted Hours', 'Allocated Hours', 'Remaining', 'Date'])

```

```python
from datetime import datetime
def convert_to_date(date_str):
    # Split the string into year and month
    year, month = date_str.split('-')
    
    # Convert the 2-digit year to 4-digit (assume 2000s)
    year = '20' + year
    
    # Parse the date with the first day of the month (default as 01)
    date = datetime.strptime(f'{year}-{month}-01', '%Y-%b-%d')
    
    # Return the formatted date (e.g., YYYY-MM-DD)
    return date.strftime('%Y-%m-%d')
```

```python
vd_all_loc1_transformed['Date'] = vd_all_loc1_transformed['Date'].apply(convert_to_date)
```
